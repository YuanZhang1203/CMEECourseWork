Working through growth rate equations


Age Specific Survivorship (l_x)

```r
M_0_par <- 10 #Arbitrary abundance value
t_par <- 10 #arbitrary time
K_par <- 100 #arbitrary carrying capacity

z_J_par <- 0.05 #1/day
z_par <- 0.03 #1/day
alp_par <- 25.0 #days, alpha
b_pk_par <- 10.0 #individuals/(individual * day)
kap_par <- 0.1 #1/day, kappa, K
```

```r
X_Age <- seq(0,60,0.1) #Age
```

```r
l_x <- exp(-z_J_par*alp_par-z_par*(X_Age-alp_par))
```

```r
plot(X_Age, l_x, type="l", xlim=(c(alp_par,60)), ylim=(c(0.1,0.3)))
```

Age Specific Fecundity (b_x)

```r
b_x <- b_pk_par*exp(kap_par*alp_par-kap_par*X_Age)
```

```r
plot(X_Age,b_x, type="l",xlim=(c(alp_par,60)), ylim=(c(0,10)))
```

Plotting l_x m_x

```r
lxbx <- l_x*b_x
```

```r
plot(X_Age,lxbx, type="l", xlim=(c(alp_par,60)), ylim=(c(0,3)))
```

Model r_m approx equation

```r
kz <- kap_par+z_par #term for k+z
```

```r
X_bpk <- seq(0.1,1000,0.1)
r_m <- (kz*(log10(X_bpk/kz)-(alp_par*z_J_par)))/(alp_par*kz +1)
```

```r
plot(X_bpk,r_m)
```

Modeling parameters onto rm equation

Sharpe schoolfield equation

```r
#arbitrary values
k_par <- 8.617*1e-5
E_par <- 0.55
E_D_par <- 4
T_pk_par <- 25.3 +273.15
T_ref_par <- 10+273.15
B_0_par <- 0.5
```

```r
T_vector <- 273.15+seq(0,40,0.05)

B_par <- B_0_par*((exp((-E_par/k_par)*((1/T_vector)-(1/T_ref_par))))/(1+((E_par/(E_D_par-E_par))*exp((E_D_par/k_par)*((1/T_pk_par)-(1/T_vector))))))
```

```r
plot(T_vector-273.15,B_par, type="l")
abline(v=25)

#plots model
```

```r
plot(T_vector-273.15,1/B_par, type="l")
abline(v=25)
#plot inverse
```

Plotting idealized Thermal performance curves

```r
T_pk_par <- 25.3+273.15
E_par <- 0.65
ED_par <- 3
k_par <- 8.617*1e-5
T_ref_par <- 10+273.15
```

```r
X_Temp <- 273.15+seq(0,40,0.05)

response <- (exp((-E_par/k_par)*((1/X_Temp)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_par)-(1/X_Temp))))
```

```r
B0_alp<- 1/alp_par 
B0_bpk<-b_pk_par 
B0_z<-1/z_par 
B0_zJ <-1/z_J_par 
B0_kap <- kap_par
```

```r
Y_alp<- 1/(response*B0_alp) 
Y_bpk<- response*B0_bpk
Y_z<- 1/(response*B0_z)
Y_zJ <-1/(response*B0_zJ)
Y_kap <- response*B0_kap
```

```r
#plots
par(mfrow=c(2,3))
plot(X_Temp-273.15,Y_alp, type="l")
plot(X_Temp-273.15,Y_bpk,type="l")
plot(X_Temp-273.15,Y_z,type="l")
plot(X_Temp-273.15,Y_zJ,type="l")
plot(X_Temp-273.15,Y_kap, type="l")
```

Find r_m (growth rate)

```r
r_m_vector <- ((Y_kap+Y_z)*(log10(Y_bpk/(Y_kap+Y_z))-(Y_alp*Y_zJ)))/((Y_alp*(Y_kap+Y_z))+1)
```

```r
plot(X_Temp-273.15,r_m_vector, ylim=c(-0.3,0.4), type="l",col="Blue")
abline(h=0,lty=2)
```

calculate niche width

```r
nicheW <- function(r_m,Temp){
    dataframe<- data.frame(Temperature=Temp,Rate=r_m)
    values <- subset(dataframe,dataframe$Rate>0)
    width<-max(values$Temperature)-min(values$Temperature)
    print(width)
}
```

```r
nicheW(r_m_vector,X_Temp)
```

Rainbow Plot : vector mistmatch and r_m response

```r
#Values
T_pk_par <- 25.3+273.15
E_par <- 0.65
ED_par <- 3
k_par <- 8.617*1e-5
T_ref_par <- 10+273.15


T_vec <- 273.15+seq(0,50,0.05)

response_mm <- (exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_par)-(1/T_vec))))

z_J_par <- 0.05 
z_par <- 0.03 
alp_par <- 25.0 
b_pk_par <- 10.0 
kap_par <- 0.1 

#B0
B0_alp<- 1/alp_par 
B0_bpk<-b_pk_par 
B0_z<-1/z_par 
B0_zJ <-1/z_J_par 
B0_kap <- kap_par

#Vectors
alp_vec<- 1/(response_mm*B0_alp)
bpk_vec<- response_mm*B0_bpk
z_vec<- 1/(response_mm*B0_z)
zJ_vec <-1/(response_mm*B0_zJ)
kap_vec <- response_mm*B0_kap

r_m_vector <- ((kap_vec+z_vec)*(log10(bpk_vec/(kap_vec+z_vec))-(alp_vec*zJ_vec)))/((alp_vec*(kap_vec+z_vec))+1)
```

```r
#set mistmatch temps, only change juvenile traits (alp,zJ)
mm <- seq(-10,10.5,0.5)
T_pk_mm <- T_pk_par + mm

plot(T_vec-273.15,r_m_vector, ylim=c(-0.5,0.5), type="l",col="blue",lwd=2)
abline(h=0,lty=2)
NicheW <-c()

for(i in 1:length(T_pk_mm)){
#these two different
alp_vec_mm <- 1/(B0_alp*((exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_mm[i])-(1/T_vec)))))) 
zJ_vec_mm <- 1/(B0_zJ*((exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_mm[i])-(1/T_vec))))))

#these are the same
bpk_vec_mm <- B0_bpk*((exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_par)-(1/T_vec)))))
kap_vec_mm <- B0_kap*((exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_par)-(1/T_vec)))))
z_vec_mm <- 1/(B0_z*((exp((-E_par/k_par)*((1/T_vec)-(1/T_ref_par))))/(1+(E_par/(ED_par-E_par))*exp((ED_par/k_par)*((1/T_pk_par)-(1/T_vec))))))    
    
    
rm_vec_mm<-((kap_vec_mm+z_vec_mm)*(log10(bpk_vec_mm/(kap_vec_mm+z_vec_mm))-(alp_vec_mm*zJ_vec_mm)))/((alp_vec_mm*(kap_vec_mm+z_vec_mm))+1)  
    
lines(T_vec-273.15,rm_vec_mm)
NicheW[[i]]<- nicheW(rm_vec_mm,T_vec) 
}

plot(mm,NicheW,type="l")
```
