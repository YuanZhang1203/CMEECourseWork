Plotting matrix of graphs with all traits

```r
library("dplyr")
data <- read.csv("TraitofInterest.csv")
```

```r
data <- data %>% mutate(Variable = case_when(originaltraitname %in% c("Development Rate", "Development Time", "Development time","Generation Time","Egg development time") ~ 'alp',
                                             originaltraitname %in% c("Fecundity","Fecundity Rate", "Oviposition Rate") ~ 'bpk',
                                             originaltraitname %in% c("Adult longevity (female, bloodfed)", "Adult longevity (male)","Adult survival","Adult survival (female, bloodfed)", "Adult survival (male)") ~ 'z',
                                            originaltraitname %in% c("Juvenile survival","Juvenile survival ") ~ 'z_J',
                                             originaltraitname %in% c("Longevity","Mortality Rate","Percentage Survival","Survival Rate","Survivorship", "Survival Time") ~ 'Mortality'))
```

```r
mort <- subset(data,data$Variable %in% c("Mortality","z","z_J"))
mort <- gdata::drop.levels(mort)
levels(mort$interactor1stage)

```

```r
#trying with Diptera first

Dip <- subset(data,data$interactor1order=="Diptera")
Dip <- gdata::drop.levels(Dip)
Diplevel <- levels(Dip$interactor1)
Diplevel
```

```r
par(mar=c(0.1,0.1,0.1,0.1))
layout(matrix(c(rep.int(1,5),2:6,rep.int(7,5),8:12, rep.int(13,5),14:18, rep.int(19,5),20:24,rep.int(25,5),26:30,rep.int(31,5),32:36),nrow=12,ncol=5,byrow=TRUE),heights=c(1,3,1,3,1,3,1,3,1,3,1,3))
layout.show(n=42)


for(i in 1:length(Diplevel)){
    species <- subset(Dip,Dip$interactor1==levels(Dip$interactor1)[i])
    species<- gdata::drop.levels(species)
    #title
    plot.new()
    text(0.5,0.5,levels(Dip$interactor1)[i],cex=2,font=2)
    
    #alpha
    alp <- subset(species,species$Variable=="alp")
    alp <- gdata::drop.levels(alp)
    if(length(levels(alp$originaltraitunit))>0){
         plot.new()
    text(0.5,0.5,"YES",cex=2,font=2)
    }else {
         plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
    }
    
      #bpk
    bpk <- subset(species,species$Variable=="bpk")
    bpk <- gdata::drop.levels(bpk)
    if(length(levels(bpk$originaltraitunit))>0){
         plot.new()
    text(0.5,0.5,"YES",cex=2,font=2)
    }else {
         plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
    }
    
    #mortality
    mortality <- subset(species,species$Variable =="Mortality")
    mortality <- gdata::drop.levels(mortality)
    
    #z
    z <- subset(species,species$Variable=="z")
    z <- gdata::drop.levels(z)
    if(length(levels(z$originaltraitunit))>0){
         plot.new()
    text(0.5,0.5,"YES",cex=2,font=2)
    }else {
        zsub <- subset(mortality,mortality$interactor1stage %in% c('Adult','adult','adult (female)','adult (male)'))
        zsub <- gdata::drop.levels(zsub)
        if(length(levels(zsub$interactor1stage))>0){
             plot.new()
    text(0.5,0.5,"YES",cex=2,font=2)
        }else {
            plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
        }
    }

    #zJ
        zJ <- subset(species,species$Variable=="z_J")
    zJ <- gdata::drop.levels(zJ)
    if(length(levels(zJ$originaltraitunit))>0){
         plot.new()
    text(0.5,0.5,"YES",cex=2,font=2)
    }else {
        zJsub <- subset(mortality,mortality$interactor1stage %in% c('1st instar','2nd instar','3rd instar','4th instar','5th instar','Egg','egg + larvae + pupae','Egg to L1','Egg-to-adult','juvenile','L1 to Adult','L1 to L2','L2 to L3','L3 to L4','L4 to Pupae','larvae','larvae + pupae','Nymph','Pupae to Adult'))
        zJsub <- gdata::drop.levels(zJsub)
        
        if(length(levels(zJsub$interactor1stage))>1){
            plot.new()
    text(0.5,0.5,list(levels(zJsub$interactor1stage)),cex=2,font=2)
        }else {
            if(length(levels(zJsub$interactor1stage))== 1){
                 plot.new()
    text(0.5,0.5,levels(zJsub$interactor1stage),cex=2,font=2)
            }else {plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
            }
        }
        }

    #Kappa
    plot.new()
    text(0.5,0.5,"<<<Refer to fecundity plot",cex=2,font=2)

}

```

```r
#removed noTime subset
plot.TPC <- function(df){
    
    time <- subset(df,df$secondstressor=="Time") 
    time <- gdata::drop.levels(time) 
if(length(unique(time$secondstressor))>0){
    
    if(length(unique(time$originaltraitunit))>1){
        plot.new()
    text(0.5,0.5,print(unique(time$originaltraitunit)),cex=2,font=2)
        
    }else{
    

if(length(unique(time$interactor1stage))>1){
#graph type 1
unqTemp <- unique(time$ambienttemp)
unqType <- levels(time$interactor1stage)
    
col<- c("Blue", "Red", "Green","gold","deeppink","tan")
pch <- c(19,17,18,15,11,8,3)

    plot(NULL, xlim=c(0,max(time$secondstressorvalue)+5), ylim=c(0,max(time$originaltraitvalue)), main=expression(paste(time$originaltraitname[1], "(TimeY+TypeY)")), xlab="Days", ylab=expression(paste(time$originaltraitname[1], "(", time$originaltraitunit[1],")")))

for(i in 1:length(unqTemp)) {
    sub <- subset(time,time$ambienttemp==unqTemp[i])
    for (j in 1:length(unqType)){
    sub1<- subset(sub,sub$originaltraitname==unqType[j])    
    points(sub1$secondstressorvalue, sub1$originaltraitvalue, pch=pch[j],col=col[i])
        legend(45,0.3, legend=c(unqTemp,unqType),title=expression(paste("Temperature(",degree,"C)")), col=c(col[1:length(unqTemp)],rep("Black",length(unqType))),pch=c(rep(15,length(unqTemp)),pch[1:length(unqType)]), cex=1)
        }
    }  
}else {
    #graph type 2
    unqTemp <- unique(time$ambienttemp)
    col<- c("Blue", "Red", "Green","gold","deeppink","tan")
    plot(NULL, xlim=c(0,max(time$secondstressorvalue)+5), ylim=c(0,max(time$originaltraitvalue)), main=expression(paste(time$originaltraitname[1])), xlab=expression(paste("Time (", time$secondstressorunit[1], ")")), ylab=expression(paste(time$originaltraitname[1], "(", time$originaltraitunit[1],")")))
    
for(i in 1:length(unqTemp)) {
    sub <- subset(time,time$ambienttemp==unqTemp[i])
    points(sub$secondstressorvalue, sub$originaltraitvalue, col=col[i], pch=19)

         }
legend(1,80, legend=unqTemp,title=expression(paste("Temperature(",degree,"C)")), col=col[1:length(unqTemp)],pch=19, cex=1.3)
     
}   
}    
}else {    
    if(length(unique(species$originaltraitunit))>1){
        plot.new()
    text(0.5,0.5,print(unique(species$originaltraitunit)),cex=2,font=2)
        
    }else{
    if(length(unique(species$interactor1stage))>1){
        #graph type 3
        col<- c("Blue", "Red", "Green","gold","deeppink","tan")
        unqType <- unique(species$interactor1stage)
        plot(NULL, main=expression(paste(species$originaltraitname[1])), xlab=expression(paste("Temperature (",degree,"C)")), ylab=expression(paste(species$originaltraitname[1], "(", species$originaltraitunit[1],")")))

    for(i in 1:length(unqType)) {
    sub <- subset(species,species$interactor1stage==unqType[i])
    points(sub$ambienttemp, sub$originaltraitvalue, col=col[i], pch=19)

         }
    legend(1,80, legend=unqType,title="Stage", col=col[1:length(unqType)],pch=19, cex=1.3)
 
        
    }else {
        #graph type 4
         plot(species$ambienttemp,species$originaltraitvalue, pch=19, main=species$originaltraitvalue[1], xlab=expression(paste("Temperature (",degree,"C)")),ylab=expression(paste(species$originaltraitname[1], "(", species$originaltraitunit[1],")")), xlim=c(0,max(species$ambienttemp, na.rm=TRUE)+5), ylim=c(0,max(species$originaltraitvalue, na.rm=TRUE)))        
    }
    }
    }   
    }
    
```

```r
Dip <- subset(data,data$interactor1order=="Diptera")
Dip <- gdata::drop.levels(Dip)
Diplevel <- levels(Dip$interactor1)

par(mar=c(0.1,0.1,0.1,0.1))
par(mfrow=c(3,5))

#Adding plots
for(i in 1:length(Diplevel)){
    species <- subset(Dip,Dip$interactor1==levels(Dip$interactor1)[i])
    species<- gdata::drop.levels(species)
    
#alpha
    alp <- subset(species,species$Variable=="alp")
    alp <- gdata::drop.levels(alp)
    if(length(levels(alp$originaltraitunit))>0){
      plot.TPC(alp)
    }else {
    plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
    }
    
      #bpk
    bpk <- subset(species,species$Variable=="bpk")
    bpk <- gdata::drop.levels(bpk)
    if(length(levels(bpk$originaltraitunit))>0){
         plot.TPC(bpk)
    }else {
         plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
    }
    
    #mortality
    mortality <- subset(species,species$Variable =="Mortality")
    mortality <- gdata::drop.levels(mortality)
    
    #z
    z <- subset(species,species$Variable=="z")
    z <- gdata::drop.levels(z)
    if(length(levels(z$originaltraitunit))>0){
         plot.TPC(z)
    }else {
        zsub <- subset(mortality,mortality$interactor1stage %in% c('Adult','adult','adult (female)','adult (male)'))
        zsub <- gdata::drop.levels(zsub)
        if(length(levels(zsub$interactor1stage))>0){
             plot.TPC(zsub)
        }else {
            plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
        }
    }

    #zJ
        zJ <- subset(species,species$Variable=="z_J")
    zJ <- gdata::drop.levels(zJ)
    if(length(levels(zJ$originaltraitunit))>0){
      plot.TPC(zJ)
    }else {
        zJsub <- subset(mortality,mortality$interactor1stage %in% c('1st instar','2nd instar','3rd instar','4th instar','5th instar','Egg','egg + larvae + pupae','Egg to L1','Egg-to-adult','juvenile','L1 to Adult','L1 to L2','L2 to L3','L3 to L4','L4 to Pupae','larvae','larvae + pupae','Nymph','Pupae to Adult'))
        zJsub <- gdata::drop.levels(zJsub)
        
        if(length(levels(zJsub$interactor1stage))>0){
            plot.TPC(zJsub)
         }
        else {plot.new()
    text(0.5,0.5,"NOPE",cex=2,font=2)
            }
        }

    #Kappa
    plot.new()
    text(0.5,0.5,"<<<Refer to fecundity plot",cex=2,font=2)

}


```
