# Culex annulirostris

```r
data <- read.csv("Fulldataset.csv")
summary <- read.csv("summary.csv")
```

```r
culex <- subset(data,data$interactor1=="Culex annulirostris")
culex <- gdata::drop.levels(culex)
culexSum <- subset(summary,summary$Species =="Culex annulirostris")
culexSum <- gdata::drop.levels(culexSum)
```

```r
levels(culexSum$Parameter)
```

# Development time

```r
aculex<- subset(culex, culex$originaltraitname %in% c("Development time","Egg development time"))
aculex <- gdata::drop.levels(aculex)
```

```r
aculex1 <- subset(aculex,aculex$originaltraitname=="Development time")
aculex1<- aculex1[!aculex1$originaltraitvalue==1000,]
aculex12 <- subset(aculex1,aculex$ambienttemp<40)

plot(aculex12$ambienttemp,pch=19,aculex12$originaltraitvalue, ylab="Generation Time (Days)", xlab=expression(paste("Temperature (",degree,"C)")), main="Devlopment Time")
```

```r
aculex2 <- subset(aculex,aculex$originaltraitname=="Egg development time")
aculex2<- aculex2[!aculex2$originaltraitvalue==1000,]

plot(aculex2$ambienttemp,aculex2$originaltraitvalue, pch=19,ylab="Generation Time (Days)", xlab=expression(paste("Temperature (",degree,"C)")), main="Egg Development Time")

```

# Fecundity

```r
Fculex <- subset(culex,culex$originaltraitname=="Fecundity")
temp <- unique(Fculex$ambienttemp)
col<- c("Blue", "Red", "Green")
plot(NULL, xlim=c(0,70), ylim=c(0,80), main="Fecundity", xlab="Days", ylab="Fecundity (eggs per female per day)")


for(i in 1:length(temp)) {
    sub <- subset(Fculex,Fculex$ambienttemp==temp[i])
    points(sub$secondstressorvalue, sub$originaltraitvalue, col=col[i], pch=19)

         }
legend(1,80, legend=temp,title=expression(paste("Temperature(",degree,"C)")), col=col,pch=19, cex=1.3)
```

```r
Fculex <- subset(culex,culex$originaltraitname=="Fecundity")
temp <- unique(Fculex$ambienttemp)

for(i in 1:length(temp)) {
    sub <- subset(Fculex,Fculex$ambienttemp==temp[i])
    plot(sub$secondstressorvalue, sub$originaltraitvalue, main=paste("Fecundity at", temp[i]," Degrees"), xlab="Days", ylab="Fecundity (eggs per female per day)")

         }
```

```r
zculex2 <- subset(zculex,zculex$originaltraitname %in% c('Adult survival (female, bloodfed)', 'Adult survival (male)'))
zculex2 <- gdata::drop.levels(zculex2)
```

```r
temp <- unique(zculex2$ambienttemp)
type <- levels(zculex2$originaltraitname)
Typelabel <- c("Male","Female (Bloodfed)")
col<- c("Blue", "Red", "Green")
pch <- c(19,17)
plot(NULL, xlim=c(0,70), ylim=c(0,1), main="Adult Mortality", xlab="Days", ylab="1-Proportion Survive")


for(i in 1:length(temp)) {
    sub <- subset(zculex2,zculex2$ambienttemp==temp[i])
    for (j in 1:length(type)){
    sub1<- subset(sub,sub$originaltraitname==type[j])    
    points(sub1$secondstressorvalue, 1-sub1$originaltraitvalue, pch=pch[j],col=col[i])

         }
    }
legend(45,0.3, legend=c(temp,Typelabel),title=expression(paste("Temperature(",degree,"C)")), col=c(col,"Black","Black"),pch=c(15,15,15,17,19), cex=1)

```

```r
zculex21 <- subset(zculex2,zculex2$originaltraitname=='Adult survival (female, bloodfed)')
temp <- unique(zculex21$ambienttemp)
col<- c("Blue", "Red", "Green")
plot(NULL, xlim=c(0,70), ylim=c(0,1), main="Adult Mortality: Female, Bloodfed", xlab="Days", ylab="1-Proportion Survive")


for(i in 1:length(temp)) {
    sub <- subset(zculex21,zculex21$ambienttemp==temp[i])
    points(sub$secondstressorvalue, 1-sub$originaltraitvalue, col=col[i], pch=19)

         }
legend(49,0.3, legend=temp,title=expression(paste("Temperature(",degree,"C)")), col=col,pch=19, cex=1.3)
```

```r
zculex22 <- subset(zculex2,zculex2$originaltraitname=='Adult survival (male)')
temp <- unique(zculex22$ambienttemp)
col<- c("Blue", "Red", "Green")
plot(NULL, xlim=c(0,70), ylim=c(0,1), main="Adult Mortality: Male", xlab="Days", ylab="1-Proportion Survive")


for(i in 1:length(temp)) {
    sub <- subset(zculex22,zculex22$ambienttemp==temp[i])
    points(sub$secondstressorvalue, 1-sub$originaltraitvalue, col=col[i], pch=19)

         }
legend(49,0.3, legend=temp,title=expression(paste("Temperature(",degree,"C)")), col=col,pch=19, cex=1.3)
```
