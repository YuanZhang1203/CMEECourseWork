## Supplementary Appendix 1: Model for the temperature-dependent fitness and density of disease vectors

In age-structured populations, the expected reproductive success of an individual can be described using the continuous form of the Euler-Lotka equation:
$$
	\int\limits_{\alpha}^\infty \mathrm{e}^{-r_m x}l_{x}b_{x}\, \mathrm{d}x = 1
$$
where $r_m$ is the population growth rate, $a$ is the age of first reproduction, $l_x$ is the age-specific survivorship, and $b_x$ is the age-specific fecundity. Solving this equation gives the population growth rate, $r_m$ as a function of life-history parameters/traits.

```python
x, l_a, b_pk, alpha, z, z_J, kappa, T, M_0, K, r_m, t = var('x l_a b_pk alpha z z_J kappa T M_0 K r_m t') #assign symbolic variables
```

As arthropods and especially disease vectors are expected to have a Type III survivorship curve, given an instantaneous mortality rate, $z$, age-specific survivorship, $l_{x}$, declines exponentially with age and can be modeled as:

\begin{equation} 
	l_{x} = l_{\alpha}\mathrm{e}^{-z(x-\alpha)}
\end{equation}

Here, $l_{\alpha}$ is the proportion of eggs surviving to adulthood (age $\alpha$), which, assuming a fixed instantaneous mortality rate across all juvenile stage, $z_J$, can be modelled as:
\begin{equation}
	l_{\alpha} = \mathrm{e}^{-\int\limits_0^\alpha z_J\, \mathrm{d}x} = 
	\mathrm{e}^{-z_J \alpha}
\end{equation}

```python
l_a = exp(-integrate(z_J,(x,0,alpha))); l_a
```

This, when substituted into the $l_x$ equation then gives

\begin{equation}
	l_x = \mathrm{e}^{-(z_J \alpha + z(x-\alpha))}
\end{equation}

The figures below show the shape of the $l_x$ model. This model can be easily adapted for a variable mortality rate, for example due to senescence, often modelled using a gompertz-type equation.

```python
l_x = l_a * exp(-z*(x - alpha)); simplify(l_x)
```

(where $x \ge \alpha $).
Next, age-specific fecundity $b_x$ is expected to reach a peak, $b_{pk}$, shortly after maturation and then decline gradually with age. We model this using a exponential relationship
$$
	b_{x} = b_{pk} \mathrm{e}^{\kappa \left(\alpha - x\right)}
$$

where we assume that fecundity peaks at the age of first reproduction, which occurs immediately at the time of adult emergence ($\alpha$), and $\kappa$ is a shape parameter that controls the spread of the fecundity schedule, that is, the rate of decline in fecundity after its peak. It is also possible that there is a systemtic temperature-dependence of the delay between $\alpha$ and the age of peak fecundity. This can be modelled using a shifted exponential or gompertz equation, but is outside the scope of the current study. The figures below show the shape of the $b_x$ model.

```python
b_x = b_pk * exp(-kappa*(x - alpha)); simplify(b_x)
# alternatives with delay (v) included: 
# bx = b_pk * exp((kappa-1)*(x-v-a)/-v) * (x-a)**(kappa-1) * v**(1-kappa) # rescaled gamma (Note that kappa > 1, always!)
# bx = b_pk * exp(-kappa*(x-v-a)); simplify(bx) # with delay till peak fecundity v
# also the shifted gompertz
```

Substituting the $l_x$ and $b_x$ models into the Euler-Lotka Equation, we get

$$	
\int\limits_{\alpha}^\infty b_{pk} \mathrm{e}^{- \alpha z_{J} + \kappa \left(\alpha - x\right) - r x + z \left(\alpha - x\right)}\, \mathrm{d}x = 1
$$

```python
EuLo = exp(-r_m * x) * l_x * b_x; simplify(EuLo)
```

And evaluating the integral produces after simplification: 

\begin{equation}
\frac{b_{pk} e^{- \alpha \left(r + z_{J}\right)}}{\kappa + r + z} = 1
\end{equation}

```python
# integrate(EuLo,(x,a,oo)) #doesn't work - parameters need to be constrained/bounded
# So integrate EuLo on sagemath using integrate(EuLo, x, a, infinity), with positivity constraints on all parameters, which gives:
EuLo_int = b_pk*exp(-alpha*z_J)/((kappa + r_m)*exp(alpha*r_m) + z*exp(alpha*r_m)); simplify(EuLo_int)
```

Solving this for $r_m$ gives:

```python
r_SP = solve(EuLo_int-1,r_m); r_SP = simplify(r_SP[0]);r_SP
```

The LambertW function part is difficult to analyze/interpret, so we approximate this by taking log of the solution to the integral above and perform a power series expansion around $r_m = 0$ of just the term $\mathrm{ln}(\kappa + r + z)$: 

```python
series(ln(r_m + kappa + z),r_m, 0,3) #
# That is, exp(log(kappa + z) + r/(kappa + z))
```

That is, 

$$
 \frac{b_{pk} e^{- \alpha z_{J}}}{z e^{\alpha r_{m}} + \left(\kappa + r_{m}\right) e^{\alpha r_{m}}}
 \approx \log{\left (\kappa + z \right )} + \frac{r_{m}}{\kappa + z} - \frac{r_{m}^{2}}{2 \kappa^{2} + 4 \kappa z + 2 z^{2}} + \mathcal{O}\left(r_{m}^{3}\right)
$$

Then, substituting the first two terms back into the integral solution and solving for $r_m$ gives:

```python
#substitute # exp(log(kappa + z) + r/(kappa + z))
EuLo_int_app = b_pk*exp(-alpha*(r_m + z_J))/ exp(log(kappa + z) + r_m/(kappa + z))

r_SP_app = solve(EuLo_int_app-1, r_m); r_SP_app = simplify(r_SP_app[0]); simplify(r_SP_app)
```

Thus the approximation is: 


\begin{equation}
	r_m \approx \frac{1}{\alpha \kappa + \alpha z + 1} \left(- \alpha \kappa z_{J} - \alpha z z_{J} + \kappa \log{\left (\frac{b_{pk}}{\kappa + z} \right )} + z \log{\left (\frac{b_{pk}}{\kappa + z} \right )}\right)  
\end{equation}

The figures below show that this approximation is very good as long as $r_m$ is small ($<1$). This range of $r_m$ values is typically where most insect growth rates lie [need references]. 

```python
from scipy.special import lambertw # need this for the lambertw

b_pk_vec = np.arange(1, 10, 0.1).astype(float) #vector of b_pk's

r_SP_app_vec = np.array([r_SP_app.evalf(subs = {b_pk:b_pkVal,z_J:z_J_par, alpha:alp_par, z:z_par, kappa:kap_par}) for b_pkVal in b_pk_vec])#use lambidy to speed up

# have to specify r_SP with lambertw:
# r_SP_vec = (-a*(kappa + z) + lambertw((a_par*b_pk_vec*exp(a_par*(kap_par + z_par - z_J_par))).astype('float')).real)/a_par
tmp = lambertw((alp_par*b_pk_vec*exp(alp_par*(kap_par + z_par - z_J_par))).astype('float')).real
r_SP_vec = -(kap_par + z_par) + tmp/alp_par

fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(r_SP_vec, r_SP_app_vec)
ax.set_xlabel('$r$ (exact)', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('$r$ (approximation)', fontsize=14, fontweight = 'bold')
plt.savefig('../results/rapprox0.pdf')
```

```python
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(r_SP_vec, r_SP_vec/r_SP_app_vec)
ax.set_xlabel('$r$ (exact)', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('Approximation as $\%$ of exact value', fontsize=14, fontweight = 'bold')
plt.savefig('../results/rapprox1.pdf')
```

Now finally, $r_m$ can be translated into population density using the logistic growth equation, which has the solution:

\begin{equation}
	M_t = \frac{M_0K\mathrm{e}^{r_mt}}{K+M_0(\mathrm{e}^{r_m t}-1)}
\end{equation}

```python
M = M_0*K*exp(r_m * t)/(K + M_0*(exp(r_m * t) -1)); M
```

## Supplementary Appendix 2: Temperature-dependence of life history parameters 

We use the Sharpe-schoolfield model: 

```python
# Assign functions
B_0, E, E_D, T_pk, k, T, T_ref = var('B_0 E E_D T_pk k T T_ref')

# B = B_0 * exp(-E/(k * T)) / (1 + (E / (E_D - E)) + exp((E_D / k)* ((1/T_pk) - (1/T)))); simplify(B)
```

```python
B = B_0 * exp(-E * ((1/(k*(T))) - (1/(k*T_ref)))) /(1 + (E / (E_D - E)) * exp((E_D / k)* ((1/T_pk) - (1/T)))); simplify(B)
```

```python
B_inv = simplify(1/B); B_inv
```

### TPC parameterizations of life history traits

A dataset of the thermal responses of vectors was collated in order to fit thermal response curves which would then be used to parametrise the model. However, the quality of the data was not sufficient to generate realiable thermal response curves.

Instead, idealised curves were generated with reference to the collated data. The Schoolfield model of thermal dependence was used as it produces a unimodal thermal response curve. Furthermore, the Schoolfield model has an explicit parameter for the temperature at which a trait is maximised, facilitating the investigation of the impact of thermal mismatches between life history traits. For development time, $a$, and adult and juvenile mortality, $z$ and $z_J$ respectively, an inverse Schoolfield model is used as the optimal trait value is a minimum value.

From the dataset it was determined that, across a range of traits, $10\mathrm{^oC}$ and $40\mathrm{^oC}$ were appropriate maximum and minimum temperatures for the thermal response curve, and that $26\mathrm{^oC}$ was an appropriate optimum temperature. The optimum temperature, $T_{pk}$, was therefore set to $26\mathrm{^oC}$. The activation energy and deactivation energy, $E$ and $E_D$ respectively, were then set as 1.7 and 4.5, respectively, such that the thermal response curves approached 0 at $10\mathrm{^oC}$ and $40\mathrm{^oC}$.

Each parameter was scaled to a maximum consistent with the data set using the scaling coefficient, $B_0$. The value assigned to each parameter are shown in the following table:

b_{pk} & 0.900 & 26 & 1.7 & 4.5 \\
z & 2.500 & 26 & 1.7 & 4.5 \\
z_J & 1.900 & 26 & 1.7 & 4.5 \\
\kappa & ? & 26 & 1.7 & 4.5 \\

\end{array}

## Supplementary Appendix 3: The temperature-dependent sensitivity analysis

### Overall sensitivity

We are interested in understanding the contribution of the TPC of each of the life-history parameters to the the TPC of population density. To this end, we write the temperature-dependent population density model as (using the chain rule): 

\begin{equation}
\frac{\mathrm{d}M}{\mathrm{d}T} = 
\frac{\partial M}{\partial b_{pk}} \frac{\mathrm{d}b_{pk}}{\mathrm{d}T} + 
\frac{\partial M}{\partial \alpha}\frac{\mathrm{d}/alpha}{\mathrm{d}T} + 
\frac{\partial M}{\partial z} \frac{\mathrm{d}z}{\mathrm{d}T} +
\frac{\partial M}{\partial z_J} \frac{\mathrm{d}z_J}{\mathrm{d}T} + 
\frac{\partial M}{\partial \kappa} \frac{\mathrm{d}\kappa}{\mathrm{d}T} 
\end{equation}
where each summed term is the relative  contribution of the parameter's TPC to the TPC of population density.

Furthermore, each of the derivatives of $M$ wrt a parameter can be written as 

$$ \frac{\partial M}{\partial b_{pk}} = \frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial b_{pk}},~\frac{\partial M}{\partial \alpha} = \frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial \alpha}, ~\mathrm{etc.}$$

These calculations are as follows.

```python
dM_dr = simplify(diff(M, r_m)); dM_dr
```

```python
dr_dbpk = diff(r_SP_app, b_pk); simplify(dr_dbpk) 
```

```python
dr_dalp = diff(r_SP_app, alpha); simplify(dr_dalp)
```

```python
dr_dz = diff(r_SP_app, z); simplify(dr_dz)
```

```python
dr_dzJ = diff(r_SP_app, z_J); simplify(dr_dzJ)
```

```python
dr_dkap = diff(r_SP_app, kappa); simplify(dr_dkap)
```

For the TPC model, we ony need to calculate the derivative wrt $T$ for either the function $B$ (for parameters that are rates) or the 1/$B$ (for parameters that are times):

```python
dB1dT =  diff(B, T); simplify(dB1dT)
```

```python
dB2dT =  diff(B_inv, T); simplify(dB2dT)
```

We can now substitute trait-specific parameterizations into the equations for bB1/dt or dB2/dt, dM_dr, and, drdb_pk, dr_da, etc to calculate the thermal parameter sensitivity of M.

Now we can evaluate 
$$
\frac{\mathrm{d}M}{\mathrm{d}T} = 
\frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial b_{pk}} \frac{\mathrm{d}b_{pk}}{\mathrm{d}T} + 
\frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial \alpha} \frac{\mathrm{d}\alpha}{\mathrm{d}T} + 
\frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial z} \frac{\mathrm{d}z}{\mathrm{d}T} +
\frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial z_J} \frac{\mathrm{d}z_J}{\mathrm{d}T} + 
\frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial \kappa} \frac{\mathrm{d}\kappa}{\mathrm{d}T} 
$$

where each summed term is the relative  contribution of the parameter's TPC to the TPC of population density.

Furthermore, each of the derivatives of $M$ wrt a parameter can be written as 

$$ \frac{\partial M}{\partial b_{pk}} = \frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial b_{pk}},~\frac{\partial M}{\partial \alpha} = \frac{\partial M}{\partial r_m} \frac{\partial r_m}{\partial \alpha}, ~\mathrm{etc.}$$

```python
#Calculate T_opt and M_80 range for each of the curves

M_T_opt = T_vec[np.argmax(M_vec)]
M_T_bpk_opt = T_vec[np.argmax(M_bpk_cons_vec)]
M_T_alp_opt = T_vec[np.argmax(M_alp_cons_vec)]
M_T_z_opt = T_vec[np.argmax(M_z_cons_vec)]
M_T_zJ_opt = T_vec[np.argmax(M_zJ_cons_vec)]
#M_T_kap_opt = T_vec[np.argmax(M_kap_cons_vec)]

print "T_opt for full model:", (M_T_opt - 273.15)
print "T_opt for constant b_pk:", (M_T_bpk_opt - 273.15)
print "T_opt for constant alpha:", (M_T_alp_opt - 273.15)
print "T_opt for constant z:", (M_T_z_opt - 273.15)
print "T_opt for constant z_J:", (M_T_zJ_opt - 273.15)
#print "T_opt for constant kappa:", (M_T_kap_opt - 273.15)

def find_M_80(TPC, T_vec):
    TPC_low = TPC[0:np.argmax(TPC)]
    TPC_high = TPC[np.argmax(TPC):len(TPC)]
    
    T_vec_low = T_vec[0:np.argmax(TPC)]
    T_vec_high = T_vec[np.argmax(TPC):len(TPC)]
    
    M_80_low = T_vec_low[(np.abs(TPC_low - (max(TPC)*0.8))).argmin()]
    M_80_high = T_vec_high[(np.abs(TPC_high - (max(TPC)*0.8))).argmin()]
    
    return [M_80_low, M_80_high]

def find_M_80_inverse(TPC, T_vec):
    
    inverse_TPC = 1/TPC
    
    TPC_low = inverse_TPC[0:np.argmax(inverse_TPC)]
    TPC_high = inverse_TPC[np.argmax(inverse_TPC):len(inverse_TPC)]
    
    T_vec_low = T_vec[0:np.argmax(inverse_TPC)]
    T_vec_high = T_vec[np.argmax(inverse_TPC):len(inverse_TPC)]
    
    M_80_low = T_vec_low[(np.abs(TPC_low - (max(inverse_TPC)*0.8))).argmin()]
    M_80_high = T_vec_high[(np.abs(TPC_high - (max(inverse_TPC)*0.8))).argmin()]
    
    return [M_80_low, M_80_high]

M_80_range = find_M_80(M_vec, T_vec)[1] - find_M_80(M_vec, T_vec)[0]
M_80_bpk_range = find_M_80(M_bpk_cons_vec, T_vec)[1] - find_M_80(M_bpk_cons_vec, T_vec)[0]
M_80_alp_range = find_M_80(M_alp_cons_vec, T_vec)[1] - find_M_80(M_alp_cons_vec, T_vec)[0]
M_80_z_range = find_M_80(M_z_cons_vec, T_vec)[1] - find_M_80(M_z_cons_vec, T_vec)[0]
M_80_zJ_range = find_M_80(M_zJ_cons_vec, T_vec)[1] - find_M_80(M_zJ_cons_vec, T_vec)[0]
#M_80_kap_range = find_M_80(M_kap_cons_vec, T_vec)[1] - find_M_80(M_kap_cons_vec, T_vec)[0]

print "M_80 range for full model:", M_80_range
print "M_80 range for constant b_pk:", M_80_bpk_range
print "M_80 range for constant alpha:", M_80_alp_range
print "M_80 range for constant z:", M_80_z_range
print "M_80 range for constant z_J:", M_80_zJ_range
#print "M_80 range for constant kappa:", M_80_kap_range
```

## Supplementary Information 4: Substitution into $R_0$

The population size of a disease vector is an important component of the basic reproductive number, $R_0$, of a vector-bourne disease.

In order to put these results in the context of epidemiology, we can substitute the thermal response curve of the popualtion size, $M$, into the $R_0$ equation adapted for malaria and mosquitoes by Mordecai $\mathrm{\textit{et al.}}$ (2013) from Dietz (1993):

```python
R_0, a, bc, EIP, PDR, N, r, M_t = var('R_0 a bc EIP PDR N r M_t') #Differentiate r_m (population growth rate) from r (human recovery rate)

R_0 = ((M_t*a**2*bc*exp(-z*EIP))/(N*r*z))**0.5; simplify(R_0) #R_0 equation from Mordecai et al. (2013)
```

where, $a$ is the biting rate of the vector, $bc$ is the vector competance, $EIP$ is the extrinsic i huncubation period of the malaria parasite, $N$ is human population density, $r$ is the rate at which humans recover from a malaria infection. As above, $z$ ($\mu$ in Mordecai $\textit{et al.}$, 2013) is the mortality rate of adult mosquitoes.

The parameters relating to humans can be assumed to be temperature independent. However, the parameters for the mosquito vector and malaria parasite will be temperature dependent. For those parameters not addressed above, thermal performance curves, fitted using either quadratic or Briere models, can be taken from Mordecai $\textit{et al.}$ (2013):

```python
c, T_m, T_0, f, g, h = var('c T_m T_0 f g h')

B_briere = c*T*(T - T_0)*(T_m - T)**0.5; B_briere

B_quadr = f*T**2 + g*T + h; B_quadr
```

```python
c_a = 0.000203
T_m_a = 42.3
T_0_a = 11.7

f_bc = -0.54
g_bc = 25.2
h_bc = -206

c_PDR = 0.000111
T_m_PDR = 34.4
T_0_PDR = 14.7

a_vec = np.array([(B_briere.evalf(subs = {c:c_a,T_m:T_m_a,T_0:T_0_a,T:temper})) for temper in (T_vec-273.15)])
bc_vec = np.array([(B_quadr.evalf(subs = {f:f_bc,g:g_bc,h:h_bc,T:temper})) for temper in T_vec-273.15])
PDR_vec = np.array([(B_briere.evalf(subs = {c:c_PDR,T_m:T_m_PDR,T_0:T_0_PDR,T:temper})) for temper in (T_vec-273.15)])

EIP_vec = (np.repeat(float('NaN'),len(T_vec)))

for i in range(len(T_vec)):
    if not (T_0_a <= T_vec[i]-273.15 <= T_m_a):
        a_vec[i] = 0
    if not (T_0_PDR <= T_vec[i]-273.15 <= T_m_PDR):
        PDR_vec[i] = 0
    else:
        EIP_vec[i] = 1/PDR_vec[i]
    if (bc_vec[i] < 0):
        bc_vec[i] = 0

        
fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(T_vec-273.15, a_vec);
ax.set_xlabel('Temperature', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('$a$', fontsize=14, fontweight = 'bold')

fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(T_vec-273.15, bc_vec);
ax.set_xlabel('Temperature', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('$bc$', fontsize=14, fontweight = 'bold')

fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(T_vec-273.15, PDR_vec);
ax.set_xlabel('Temperature', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('$PDR$', fontsize=14, fontweight = 'bold')
```

Substituting these TPCs, and the curves for $M$ and $z$ above, into the $R_0$ equation:

```python
N_par = 1000
r_par = 0.25

R_0_vec = np.array([(R_0.evalf(subs = {M_t:M_vec[i],a:a_vec[i],bc:bc_vec[i],z:z_vec[i],EIP:EIP_vec[i],N:N_par,r:r_par})) for i in range(len(T_vec))])

fig = plt.figure(); ax = fig.add_subplot(111)
ax.plot(T_vec-273.15, R_0_vec);
ax.set_xlabel('Temperature', fontsize=14, fontweight = 'bold'); 
ax.set_ylabel('$R_0$', fontsize=14, fontweight = 'bold')
```
